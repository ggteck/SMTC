```python
#%%
import os
from datetime import datetime, timedelta, timezone
import win32com.client

output_folder='.'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Connect to Outlook
outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")

# Access the Inbox
inbox = outlook.GetDefaultFolder(6)  # 6 refers to the Inbox folder

# Define the date range (last 7 days)
date_limit = datetime.now(timezone.utc) - timedelta(days=1)

# Access messages in the Inbox
messages = inbox.Items
messages.Sort("[ReceivedTime]", True)  # Sort messages by received time (descending)
#%%

#%%
for message in messages:
    # Ensure we only process emails received after the date limit
    received_time = message.ReceivedTime
    received_time = datetime.fromtimestamp(received_time.timestamp(), timezone.utc)
    if received_time < date_limit:
        break  # Stop processing older emails
    
    print(f"Processing email: {message.Subject} (Received: {received_time})")
    
    # Check if the email has attachments
    if message.Attachments.Count > 0:
        for attachment in message.Attachments:
            # Save the attachment
            attachment_name = attachment.FileName
            attachment_path = os.path.join(output_folder, attachment_name)
            attachment.SaveAsFile(attachment_path)
            print(f"Saved attachment: {attachment_name} to {attachment_path}")
#%%
attachment_path
```
