```python
# Catalogo
from github import Github
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML, Markdown
from dotenv import load_dotenv
import os
import base64
load_dotenv(override=True)
token = os.getenv('ITHUB_TOKEN')

def show_popup_message(message):
    display(Markdown(f"### **{message}**"))

def is_file_open(filepath):
    if not os.path.exists(filepath):
        return False  # File does not exist
    try:
        with open(filepath, 'a'):
            pass
        return False  # File is not open
    except PermissionError:
        return True 
    
def display_dataframe(df):
    header_widgets = [
        widgets.HTML(value=f"<b>{col}</b>", 
                     layout=widgets.Layout(border="1px solid black", background="lightgrey", padding="5px", margin="20px"))
        for col in ['Notebook','Documentacion','Versiones']
    ]
    
    cell_widgets = [
        df.iloc[i, j] if isinstance(df.iloc[i, j], widgets.Widget) else widgets.HTML(value=str(df.iloc[i, j]), layout=widgets.Layout(border="1px solid black", padding="5px", margin="0px")) if j == 0 else widgets.HTML(
            value=df.iloc[i, j].replace('\n', '<br>'), 
            layout=widgets.Layout(border="1px solid black", padding="5px", margin="0px")
        )
        for i in range(len(df)) for j in range(3)
    ]
    
    table = widgets.GridBox(
        children=header_widgets + cell_widgets,
        layout=widgets.Layout(
            grid_template_columns="30% 40% 30%",
            grid_gap="5px"
        )
    )
    display(table)

if token is None:
    with open('env') as f:
        token = f.read().strip()
if token is None:
    raise ValueError("GitHub token not found. Please ensure it is set in the .env file.")

g = Github(token)
repo_name = "ggteck/SMTC"
repo = g.get_repo(repo_name)

contents = repo.get_contents("")

scripts = []
while contents:
    file_content = contents.pop(0)
    if file_content.type == "dir":
        contents.extend(repo.get_contents(file_content.path))
    else:
        if (file_content.path.endswith('.ipynb'))|(file_content.path.endswith('.html')):
            scripts.append(file_content)

data = []
for script in scripts:
    commits = repo.get_commits(path=script.path)
    commit_info = [commit.commit.message.split('\n', 1)[0] for commit in commits]
    if len(commit_info)>3:
        commit_info=commit_info[0:3]
    data.append({
        'Script Name': script.path,
        'Version History': '\n'.join(commit_info),
        'Script Path': script.path
    })

df = pd.DataFrame(data)

def create_download_button(script_path):
    import requests

    script_name = script_path.split('/')[-1]
    button = widgets.Button(description=script_name, layout=widgets.Layout(width='100%', border='1px solid black'))

    def on_click(b):
        try:
            # Construct the API URL for fetching raw content
            raw_url = f"https://api.github.com/repos/{repo_name}/contents/{script_path}"

            # Request file metadata (includes download URL)
            headers = {"Authorization": f"token {token}"}
            response = requests.get(raw_url, headers=headers)
            response.raise_for_status()
            file_data = response.json()

            if "download_url" in file_data:
                # Use the direct download URL
                download_url = file_data["download_url"]

                # Fetch the actual file content
                file_response = requests.get(download_url, headers=headers)
                file_response.raise_for_status()
                content = file_response.text

                filename = script_name
                if is_file_open(filename):
                    show_popup_message(f"Error: Favor de cerrar el archivo. {filename}")
                    return

                with open(filename, 'w', encoding='utf-8') as f:
                    f.write(content)

                print(f"Downloaded {filename}")
            else:
                show_popup_message(f"Error: Unable to retrieve download URL for {script_name}")

        except Exception as e:
            show_popup_message(f"Error downloading {script_name}: {str(e)}")
    button.on_click(on_click)
    container = widgets.HBox([button], layout=widgets.Layout(border="1px solid black", padding="5px", margin="0px", width="100%"))
    return container

df['Script Name'] = df['Script Name'].apply(create_download_button)
df['Process Name']=df['Script Path'].str.split('/',expand=True)[1]
df['Process Name']=df['Process Name'].str.split('.',expand=True)[0]
df_doc=df[df['Script Path'].str.endswith('html')].copy()
df_doc.rename({'Script Name':'Documentation'},axis=1,inplace=True)
df_scripts=df[~df['Script Path'].str.endswith('html')]
df=df_scripts.merge(df_doc[['Process Name','Documentation']],how='left',on='Process Name')
df=df[['Script Name','Documentation','Version History']]
df.fillna('',inplace=True)


display_dataframe(df)
```
